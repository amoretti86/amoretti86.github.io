## Part 0: Board Representation (Provided)

We'll represent the board as a list of 9 elements:
- `'X'` for the maximizing player (AI)
- `'O'` for the minimizing player (human)
- `' '` for empty space

```
Board indices:
0 | 1 | 2
---------
3 | 4 | 5
---------
6 | 7 | 8
```

````python
def create_board():
    """Create an empty tic-tac-toe board."""
    return [' '] * 9

def print_board(board):
    """Display the board in a readable format."""
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("-----------")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("-----------")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print()

def get_available_moves(board):
    """Return list of indices where moves can be made."""
    return [i for i in range(9) if board[i] == ' ']

def make_move(board, position, player):
    """Place a player's mark at the given position. Returns new board."""
    new_board = board.copy()
    new_board[position] = player
    return new_board

# Test the board functions
test_board = create_board()
test_board = make_move(test_board, 4, 'X')
test_board = make_move(test_board, 0, 'O')
print("Test board:")
print_board(test_board)
print("Available moves:", get_available_moves(test_board))
````

**Key Idea:**
- **Alpha:** Best score maximizer can guarantee (starts at -∞)
- **Beta:** Best score minimizer can guarantee (starts at +∞)
- **Prune when:** Alpha ≥ Beta (no need to explore further)

```

    3   5   
```
