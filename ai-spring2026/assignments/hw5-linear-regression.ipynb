## Part 5 — Gradient Descent

The closed-form solution is elegant, but in practice many models are too complex for a direct solution. **Gradient descent** iteratively moves parameters in the direction of steepest descent of the loss.

For MSE loss $L = \frac{1}{N}\sum(y_i - (mx_i+b))^2$, the gradients are:

$$\frac{\partial L}{\partial m} = -\frac{2}{N}\sum_{i=1}^{N} x_i(y_i - \hat{y}_i)$$

$$\frac{\partial L}{\partial b} = -\frac{2}{N}\sum_{i=1}^{N}(y_i - \hat{y}_i)$$

The update rule at each iteration is:

$$m \leftarrow m - \alpha \frac{\partial L}{\partial m}, \qquad b \leftarrow b - \alpha \frac{\partial L}{\partial b}$$

where $\alpha$ is the **learning rate**.

The **coefficient of determination** $R^2$ measures how much of the variance in $y$ is explained by the model:

$$R^2 = 1 - \frac{\text{SSE}}{\text{SST}} = 1 - \frac{\sum_i (y_i - \hat{y}_i)^2}{\sum_i (y_i - \bar{y})^2}$$

## Part 9 — Cross-Validation for Degree Selection
